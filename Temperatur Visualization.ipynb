```python
import pandas as pd
from datetime import datetime

#read CSV File
weather = pd.read_csv('weather_data.csv')
#Convert timestamp to datetime and format
weather['utc_timestamp'] = pd.to_datetime(weather['utc_timestamp'],format='%Y-%m-%d')
weather['utc_timestamp'] = weather['utc_timestamp'].dt.strftime("%d-%m-%Y")
#create table
w=weather[['utc_timestamp','DE_temperature','DE_windspeed_10m','DE11_temperature','DE11_windspeed_10m']]
w.shape
```

```python
#index
w = w.set_index('utc_timestamp')


```

```python
#last 10 tuples of the table 
w.tail(10)
```

```python
#draw plot

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(11, 4)})
w['DE_temperature'].plot(linewidth=0.5);
```

```python
#draw multiple plots
cols_plot = ['DE_temperature', 'DE11_temperature']
axes = w[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('total Temp')
```

```python
#draw plot for given range
ax = w.loc['01-01-2016':'31-12-2016', 'DE_temperature'].plot()
ax.set_ylabel('total Temp');
```

```python
#plot 
import matplotlib.dates as mdates
fig, ax = plt.subplots()
ax.plot(w.loc['01-01-2016':'10-01-2016', 'DE_temperature'], marker='o', linestyle='-')
ax.set_ylabel('total Temp')
ax.set_title('Jan-April 2016 Temperatur')
```
